```python
# As there are arund 150 columns, but we want few of them for our analysis, we will select the needed columns.
df = df[['iyear','imonth','iday','country_txt','provstate','region_txt','city','natlty1_txt','latitude','longitude','nkill','nwound','suicide',
 'attacktype1_txt','gname','targtype1_txt','weaptype1_txt','weaptype1','motive']]
```

```python
# Renaming the columns which we have selected from 150 columns.
df.rename(columns={'iyear':'Year','imonth':'Month',
 'iday':'Day',
 'country_txt':'Country','provstate':'State',
 'region_txt':'Region','city':'City',
 'latitude':'Latitude',
 'longitude':'Longitude','natlty1_txt':'Nationality','nkill':'Killed','nwound':'Wounded','suicide':'Sucide',
 'attacktype1_txt':'Attack Type','gname':'Group_Name','targtype1_txt':'Target_Type','weaptype1_txt':'Weapon_Type','weaptype1':'Weapon','motive':'Motive'},inplace = True)
```

```python
# Chech the new dataset which is created from the selected values.
df
```

```python
# Now, lets check the information of the dataset like the datatype of the values, total values, not null values.
df.info()
```

```python
# Where ever the value will be null, it will show the TRUE otherwise it will show the FALSE.
df.isnull()
```

```python
# Now, lets chech the total sum of nunn values of each columns.
df.isnull().sum()
```

```python
corelation = df.corr()
corelation
```

```python
# Here we are filling the null values with '0' and setting its datatype to int
df['Wounded'] = df['Wounded'].fillna(0).astype(int)
df['Killed'] = df['Killed'].fillna(0).astype(int)
```

```python
# We created a new Column whose name is Casualities by concatinating two Columns Wounded and Killed.
df['Casualities'] = df['Wounded'] + df['Killed']
```
